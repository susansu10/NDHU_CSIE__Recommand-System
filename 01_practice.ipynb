### Meachine Learning 是甚麼？
> 計算機從數據中學習並完成任務，例如垃圾郵件過濾器

### 為何用機器學習
- 用數據訓練演算法，更快、更容易維護程序，也更準確
- 可以解決超級複雜問題
- 在不斷變化的環境中容易自動重新訓練
- 可能可以提供有關 ML 問題的解決方法見解

### 機器學習種類
- 是否受過人工監督訓練 (監督、半監督、無監督、強化學習)
- 是否可以即時增量學習
- 是否將新數據點和已知數據點進行比較來工作，或者檢測數據中的模式並構建預測模型 (基於實例與基於模型的學習)

### 監督式學習
> 給定標記數據並進行分類或迴歸

### 無監督式學習
> 無標籤模式識別如分群或異常偵測

### 半監督式學習
> 使用部分標記的數據解決監督學習問題

### 強化學習
> 訓練機構選擇行動以最大化數字獎勵公制

### 批次 / 離線學習
- 無法逐步學習
- 必須使用所有可行數據進行訓練
- 為了適應新數據，新版本的系統應該在完整數據集上重新訓練

### 增量 / 線上學習
 - 通過向系統提供數據實例來單獨逐步訓練系統
 - 設置學習速率
 
### 實例式學習
> 牢記範例，通過使用相似性度量將它們與學習過的範例進行比較並歸納成新案例

### 模型式學習
- 構建這些範例的模型，然後使用該模型來預測
- 模型選擇、性能測量 (適應或代價函數)

### 機器學習挑戰 : 不好的數據
- 訓練數據量不足 (也許數據比演算法更重要)
- 非代表性訓練數據 (源自採樣噪音或偏差)
- 質量差的數據 (充滿錯誤、異常值和噪音)
- 不相關的特徵 (特徵選擇、特徵提取 : 結合現有特徵產生一個更有用的)

### 機器學習挑戰 : 不好的演算法
- 過度擬合 / 低度擬合 訓練的數據

### 測試與驗證
- 將數據分成兩組：訓練集和測試集
- 在訓練集中訓練模型並在測試中評估模型
- 新案例的錯誤率稱為泛化錯誤（泛化錯誤差)
- 如果訓練誤差低，但泛化誤差高，則意味著過度擬合
- 比較不同的性能關於測試集上的超參數調整和模型選擇有時是不好的
- 一種解決方案是基於保留驗證
- 如果驗證集太小，我們可以應用交叉驗證
